```python
import csv

# Esta funcion devuelve un diccionario de diccionarios llamado "stationMagnitude".
# Este contiene para cada estacion (key) el total acumulado (value) de cada magnitud (sub-key).

def obtainStationMagnitude(magnitudeName):

    # Apertura del fichero.
    
    csvFile = open('Contaminación.csv',encoding="utf8",errors='ignore');

    # Objeto lector que separa las lecturas por ";".
    
    csvReader = csv.reader(csvFile,delimiter=";");

    # Esta variable contendra para cada estacion, cuanta cantidad hay de cada magnitud.
    
    stationMagnitude = dict();

    # Contiene el indice de la columna del campo MAGNITUD.
    
    magnitudeColumn = -1;

    # Contiene el indice de la columna del campo ESTACION.
    
    stationColumn = -1

    # Recorremos todas las filas del fichero de contaminacion.
    
    for row in csvReader:

        # Si estamos en la fila de la cabecera.
        
        if(csvReader.line_num == 1):

            # Recorremos todas las columnas y nos quedamos con los indices de las columnas ESTACION y MAGNITUD.
            
            for column in range(0,len(row)):

                if(row[column] == "ESTACION"):

                    stationColumn = column;

                elif(row[column] == "MAGNITUD"):

                    magnitudeColumn = column;

            if(stationColumn == -1 or magnitudeColumn == -1):
                
                raise Exception("ERROR: No se han detectado todas las siguientes columnas, ESTACION, MAGNITUD.");
                    
        else:

            # Obtenemos los codigos de la estacion y la magnitud.
            
            station = int(row[stationColumn]);
            magnitude = int(row[magnitudeColumn]);

            # Si aun no tenemos registrada esta estacion la añadimos.
            
            if(not station in stationMagnitude.keys()):

                # Cada estacion tiene mediciones de una o varias magnitudes.
                
                stationMagnitude[station] = dict();

                # Inicializamos todas las magnitudes que puede medir una estacion.
                
                for m in magnitudeName.keys():

                    stationMagnitude[station][m] = 0;

            # Recorremos las columnas que poseen la cantidad de una magnitud y si son validas las acumulamos.
                    
            for j in range(7,len(row),2):

                if(row[j+1] == 'V'):

                    stationMagnitude[station][magnitude] += float(row[j]);

    csvFile.close();

    return stationMagnitude;

# En esta funcion mostraremos la tabla y grabaremos el fichero 'Salida.cvs'.

def showTableAndWriteFileCsv(stationMagnitude,stationName,magnitudeName):

    # Abrimos el fichero en modo escritura.
    
    csvFile = open('Salida.csv','w',encoding="utf8");

    # Cargamos un objeto escritor.
    
    csvWriter = csv.writer(csvFile);

    # Recorremos todas las estaciones ordenadas por su codigo de estacion de menor a mayor.
    
    for station in sorted(stationMagnitude.keys()):

        # Esta variable contendra el registro a grabar en el fichero 'Salida.csv'.
        
        row = [stationName[station]];

        print("{:*^72s}".format(stationName[station]));

        # Recorremos las magnitues ordenadas por su codigo de  magnitud de menor a mayor.
        # Añadimos al registro a grabar el nombre de la magnitud seguido de su valor acumulado.
        
        for magnitude in sorted(stationMagnitude[station].keys()):
            
            row.append(magnitudeName[magnitude]);
            row.append(stationMagnitude[station][magnitude]);

            print("{:*^36s}".format(magnitudeName[magnitude]),end = ''); 
            print("{:*^36f}".format(stationMagnitude[station][magnitude],2));

        print(""); 

        # Escribimos el registro en el fichero 'Salida.csv'.
        
        csvWriter.writerow(row);

    # Cierre del fichero y volcado de cambios.
        
    csvFile.close();

#**********************************************************************************************************************

try:

    # Los nombres de las magnitudes son los mismos quelos existentes en el fichero PDF en la sección ANEXO II.
    
    magnitudeName = {
        1:"Dióxido de Azufre",
        6:"Monóxido de Carbono",
        7:"Monóxido de Nitrógeno",
        8:"Dióxido de Nitrógeno",
        9:"Partículas < 2.5 μm",
        10:"Partículas < 10 μm",
        12:"Óxidos de Nitrógeno",
        14:"Ozono",
        20:"Tolueno",
        30:"Benceno",
        35:"Etilbenceno",
        37:"Metaxileno",
        38:"Paraxileno",
        39:"Ortoxileno",
        42:"Hidrocarburos totales (hexano)",
        43:"Metano",
        44:"Hidrocarburos nometánicos (hexano)"
    }

    # Los nombres de las estaciones han de ser los mismos que los existentes en el fichero 'Estaciones.csv'.
        
    stationName = {
        1:"Pº. Recoletos",
        2:"Glta. de Carlos V",
        3:"Pza. del Carmen",
        35:"Pza. del Carmen",
        4:"Pza. de España",
        5:"Barrio del Pilar",
        39:"Barrio del Pilar",
        6:"Pza. Dr. Marañón",
        7:"Pza. M. de Salamanca",
        8:"Escuelas Aguirre",
        9:"Pza. Luca de Tena",
        10:"Cuatro Caminos",
        38:"Cuatro Caminos",
        11:"Avda. Ramón y Cajal",
        12:"Pza. Manuel Becerra",
        13:"Vallecas",
        40:"Vallecas",
        14:"Pza. Fdez. Ladreda",
        15:"Pza. Castilla",
        16:"Arturo Soria",
        17:"Villaverde Alto",
        18:"C/ Farolillo",
        19:"Huerta Castañeda",
        20:"Moratalaz",
        36:"Moratalaz",
        21:"Pza. Cristo Rey",
        22:"Pº. Pontones",
        23:"Final C/ Alcalá",
        24:"Casa de Campo",
        25:"Santa Eugenia",
        26:"Urb. Embajada (Barajas)", 
        27:"Barajas",
        47:"Méndez Alvaro", 
        48:"Pº Castellana",
        49:"Retiro",
        50:"Pza. Castilla",
        54:"Ensanche Vallecas",
        55:"Urb. Embajada (Barajas)",
        56:"Plaza Elíptica",
        57:"Sanchinarro",
        58:"El Pardo",
        59:"Parque Juan Carlos I",
        60:"Tres Olivos",
        86:"Tres Olivos"
    }

    # Aqui obtenemos para toda estacion el total acumulado de cada magnitud.

    stationMagnitude = obtainStationMagnitude(magnitudeName);

    # Muestra la tabla y graba el fichero 'Salida.csv'.

    showTableAndWriteFileCsv(stationMagnitude,stationName,magnitudeName);

except Exception as err:
    
    print(err.args);
```

```python
import csv
import json
import math

# Funcion que calcula las distancias dadas dos latitudes y longitudes.

def haversine(lat1, lon1, lat2, lon2):
    rad=math.pi/180
    dlat=lat2-lat1
    dlon=lon2-lon1
    R=6372.795477598
    a=(math.sin(rad*dlat/2))**2 + math.cos(rad*lat1)*math.cos(rad*lat2)*(math.sin(rad*dlon/2))**2
    distancia=2*R*math.asin(math.sqrt(a))
    return distancia

# Esta funcion va a devolver dos objetos:
# Una lista de listas "museumNameAndLocation" que tiene los nombres y localizaciones de todos los museos.
# Un diccionario "museumNameAndLocation" que se usara mas adelante, pero por ahora solo tendra cargadas las claves
# de todos los museos existentes.

def obtainMuseumsNamesAndLocations():

    # Esta variable es una lista de listas que contendra todos los nombres y localizaciones de cada museo.
    
    museumNameAndLocation = []; 

    # Esta variable tiene como primera clave el nombre de un museo, como sub-clave la distancia
    # existente hasta una estacion y como valor el nombre de la estacion que se encuentra a dicha distancia.
    
    museumDistanceStation = dict();

    # Cargamos el documento Json en una variable.
    
    jsonDocument = json.loads(open('Museos.json',encoding="utf8",errors='ignore').read())

    # Con este bloque comprobaremos si se encuentra la clave que nos dara acceso a los sub-documentos: "@graph".
    
    for key in jsonDocument:

        if(key == "@graph"):
            
            subDocumentsKey = "@graph";

    if(subDocumentsKey != "@graph"):
        
        raise Exception("ERROR: No se puede acceder a los sub-documentos, tal vez el documento haya cambiado.");
            
    # Recorremos todos los documentos contenidos en la clave "@graph".
        
    for document in jsonDocument[subDocumentsKey]:

        # Si existe la clave title, entonces podemos acceder al nombre del museo, de lo contrario no lo contabilizamos.
        
        if("title" in document.keys()):
            
            # Lista que contendra el nombre del museo, su latitud y su longitud.

            information = [document["title"]];

            # Esta variable se usara mas adelante, por ahora almacenara como claves los nombres de los museos.

            museumDistanceStation[document["title"]] = dict();

            # Comprobamos que el documento posea el campo "location", un ejemplo que no lo tiene es "Museo de Mineralogía",
            # en este caso introducimos NONE indicando que hay ausencia de informacion.

            if("location" in document.keys()):

                information.append(document["location"]["latitude"]);
                information.append(document["location"]["longitude"]);

            else:

                information.append(None);
                information.append(None);

            # Esta variable es una lista de listas que contiene todos los nombres y localizaciones de cada museo.

            museumNameAndLocation.append(information);
        
    return museumNameAndLocation,museumDistanceStation;
    
# Devolvemos una lista de listas que contendra todos los nombres y ubicaciones de las estaciones.
            
def obtainStationLongitudeLatitude():

    # Apertura del fichero 'Estaciones.csv'.
    
    csvFile = open('Estaciones.csv',encoding="utf8",errors='ignore');

    # Creamos un objeto lector para documentos CSV, indicando que en este caso los campos se  separan por CSV.
    
    csvReader = csv.reader(csvFile, delimiter=",");
    
    # Declaramos variables para almacenar el valor de la columna (indice) en el que se encuentra cada campo.
    
    stationColumn = -1;
    latitudeColumn = -1;
    longitudeColumn = -1;
    
    # Lista de listas que contendra todos los nombres y ubicaciones de todas las estaciones.
            
    stationLongitudeLatitude = [];
    
    for row in csvReader:

        # Usaremos la fila de la cabecera para saber si el fichero no ha sufrido modificaciones.
        
        if(csvReader.line_num == 1):

            # Recorremos todas las columnas para saber ubicar cada elemento dentro del fichero.
            
            for column in range(0,len(row)):

                if(row[column] == "ESTACION"):

                    stationColumn = column;

                elif(row[column] == "LATITUD"):

                    latitudeColumn = column;

                elif(row[column] == "LONGITUD"):

                    longitudeColumn = column;

            if(stationColumn == -1 or latitudeColumn == -1 or longitudeColumn == -1):
                
                raise Exception(stationColumn,latitudeColumn,longitudeColumn,"ERROR: No se han localizado todas las siguientes columnas, ESTACION, LATITUD, LONGITUD.");
                    
        else:

            # Grabamos los datos en una lista de listas que contendra todos los nombres y ubicaciones de las estaciones.
            
            stationLongitudeLatitude.append([row[stationColumn],float(row[latitudeColumn]),float(row[longitudeColumn])]);

    # Cerramos el fichero.
    
    csvFile.close();
    
    return stationLongitudeLatitude;

# Esta funcion devuelve un diccionario que contiene como primera clave un nombre de museo, como sub-clave
# el valor de la distancia que la separa de una estacion concreta y finalmente el valor es dicha estacion.

def obtainMuseumDistanceStation(stationLongitudeLatitude,museumNameAndLocation,museumDistanceStation):
    
    # Recorremos la informacion de cada museo almacenada en una lista de listas.
    
    for row in museumNameAndLocation:
    
        # Recorremos la informacion de cada estacion almacenada en una lista de listas.
    
        for r in stationLongitudeLatitude:

            # Si tanto el museo como las estaciones tienen latitud como longitud, entonces calculamos la distancia.
            
            if((row[1] != None) and (row[2] != None) and (r[1] != None) and (r[2] != None)):

                # Distancia entre el museo y la estacion correspondiente.
                
                distance = haversine(row[1],row[2],r[1],r[2]);

                # Lo guardamos todo en un diccionario de la siguiente forma.
                
                museumDistanceStation[row[0]][distance] = r[0];
                
    return museumDistanceStation;

# Esta funcion devuelve diccionario de diccionarios que tiene por primera clave el nombre de una estacion,
# como sub-clave la cantidad de una magnitud y como valor el nombre de dicha magnitud.

def obtainStationMaxQuantityPollutantType():

    # Abrimos el fichero 'Salida.csv'.
    
    csvFile = open('Salida.csv',encoding="utf8",errors='ignore'); # newline=""

    # Creamos un objeto lector del fichero CSV.
    
    csvReader = csv.reader(csvFile);

    stationMaxQuantityPollutantType = dict();
    
    # Recorremos cada fila del fichero 'Salida.csv'.
    
    for row in csvReader:
    
        if(row == []):
            
            continue;
    
        # Preparamos un diccionario de diccionarios que tiene por primera clave el nombre de una estacion,
        # como sub-clave la cantidad de una magnitud y como valor el nombre de dicha magnitud.
        
        stationMaxQuantityPollutantType[row[0]] = dict();
        
        # Recorremos de 2 en 2 las columnas del fichero.
        # La columna 0 es el nombre de la estacion, por eso hemos de saltarnosla.
        # La columna "column" es el nombre de una magnitud y "column+1" el valor que la magnitud toma.
        
        for column in range(1,len(row),2):
            
            stationMaxQuantityPollutantType[row[0]][float(row[column+1])] = row[column];
            
    # Cerramos el fichero.
            
    csvFile.close();
    
    return stationMaxQuantityPollutantType;
    
def showTableAndWriteFileJson(stationMaxQuantityPollutantType,museumDistanceStation):
    
    # Creamos un diccionario, este servira para grabar mas adelante el documento json.
    
    mainDoc = {
      "año": 2019,
      "fuente": "Ayuntamiento de Madrid",
      "museos": []
    };

    # Abrimos el fichero 'Salida.json'.
    
    jsonFile = open('Salida.json','w',encoding="utf8");

    # Recorremos todos los nombres de los museos.
    
    for museum in museumDistanceStation.keys():

        # Creamos un diccionario para generar un sub-documento.
        
        museosInfo = dict();
        
        # Nombre del primer campo del sub-documento.

        museosInfo["Museo"] = museum;

        # Añadimos el sub-documento a la lista del campo "museos" del documento principal.
        
        mainDoc["museos"].append(museosInfo);

        # Creamos una lista orenada de menor a mayor de las distancias que separan al museo de las estaciones.
        
        distances = sorted(museumDistanceStation[museum]);

        print("{:*^72s}".format(museum));

        # Grabamos un maximo de 3 estaciones y un minimo de cero. 
            
        for i in range(0,3 if len(museumDistanceStation[museum]) >= 3 else len(museumDistanceStation[museum])):

            # Obtenemos el nombre de la 1,2,3... estacion mas cercana.
            
            station = museumDistanceStation[museum][distances[i]];

            # Obtenemos la cantidad de polucion mas alta de la la estacion especificada.
            
            quantities = sorted(stationMaxQuantityPollutantType[station].keys(),reverse=True);

            # Dentro del campo "museos" hay una lista, con -1 accedo al ultimo elemento de ella para poder
            # crear diccionarios (documentos Json) dentro de ella.
            
            mainDoc["museos"][-1][f"Estación {i+1}"] = {"Nombre":station,"Valor":quantities[0],"Tipo":stationMaxQuantityPollutantType[station][quantities[0]]};

            print("{:*^36f}".format(quantities[0]),end='');
                    
            print("{:*^36s}".format(stationMaxQuantityPollutantType[station][quantities[0]]));
            
        print("");

    # Escribo el documento Json sin limitar al codigo ascii ya que este no reconoce los simbolos no definidos en el.
        
    jsonFile.write(json.dumps(mainDoc,ensure_ascii=False));

    # Cerramos el fichero y volcamos los cambios.
    
    jsonFile.close();

#***********************************************************************************************************************

try:

    # Obtenemos todos los museos y sus ubicaciones.
    
    museumNameAndLocation , museumDistanceStation = obtainMuseumsNamesAndLocations();

    # Obtenemos todas las estaciones y sus ubicaciones.
    
    stationLongitudeLatitude = obtainStationLongitudeLatitude();

    # Obtenemos para cada museo la distancia existente a cada estacion.
    
    museumDistanceStation = obtainMuseumDistanceStation(stationLongitudeLatitude,museumNameAndLocation,museumDistanceStation);

    # Obtenemos para cada estacion, las cifras (de mayor a menor) de contaminacion de cada tipo de contaminante.
    
    stationMaxQuantityPollutantType = obtainStationMaxQuantityPollutantType();

    # Muestra la tabla por pantalla y escribe el fichero Json.
    
    showTableAndWriteFileJson(stationMaxQuantityPollutantType,museumDistanceStation);

except Exception as err:
    
    print(err.args);
```
